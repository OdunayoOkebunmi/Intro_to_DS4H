Numpy (and matplotlib): artificial generation of a databse

```python
import numpy as np
```

```python
y1 = 0
y2 = 1
N1 = 200
N2 = 200
d = 1
sigma1 = 0.5
sigma2 = 0.5
```

```python
x1 = (d/2)*np.array([-1,-1])
x2 = (d/2)*np.array([1,1])
X_T=(d/2)*np.array([[ -1. ,1.],[-1.,1.]])
X = np.transpose(X_T)

print(np.shape(X_T))
print(len(X_T))
print(np.size(X_T))
print(X[1,1])
```

```python
mu1, sigma_1 = x1, sigma1
mu2, sigma_2 = x2, sigma2
X1=np.random.normal(mu1,sigma1, size=(N1-1,2))
X2=np.random.normal(mu2,sigma2, size=(N2-1,2))
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
plt.scatter(X1[:,0],X1[:,1]);
plt.scatter(X2[:,0],X2[:,1], c='r');
plt.title('Scatter plot of the classes');
plt.xlabel('$\mathbf{x}_1$');
plt.ylabel('$\mathbf{x}_2$');
```

```python
X = np.vstack((x1,x2,X1, X2))
plt.scatter(X[:,0],X[:,1]);
plt.title('Scatter plot of the classes');
plt.xlabel('$\mathbf{x}_1$');
plt.ylabel('$\mathbf{x}_2$');
y = np.array([1]*200 + [0]*200)
print(y.shape,X.shape)
```

```python
from sklearn import linear_model as lm
from sklearn import metrics as met

logreg = lm.LogisticRegression()
logreg.fit(X,y)
yhat = logreg.predict(X)
# print(yhat)
# np.column_stack((y,yhat))
accuracy_score = met.accuracy_score(y,yhat)
print ( "The classifier offers a good prediction in % s %% cases." % ( 100 * accuracy_score )) 
```

```python
from sklearn.cluster import KMeans
from sklearn import metrics as met
kmeans = KMeans(n_clusters = 2)
kmeans.fit(X,y)
yhat = kmeans.predict(X)
# print(yhat)
# np.column_stack((y,yhat))
accuracy_score = met.accuracy_score(y,yhat)
print ( "The classifier offers a good prediction in % s %% cases." % ( 100 * accuracy_score )) 
```
